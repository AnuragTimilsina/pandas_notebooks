```python
import pandas as pd
df = pd.read_csv("./data/weather_data.csv")
df
```

```python
df = df[9:]
df
```

```python
df = df.reset_index()
df
```

```python
df.drop(['index'], axis=1, inplace=True)
```

```python
df
```

```python
df.set_index('location', inplace=True)
```

# Fill NA:

```python
df = df.fillna(0)
```

```python
new_df = df.fillna({
    'location': 0,
    'Kathmandu': 0,
    'Kathmandu.2': 'none',
})
new_df
```

```python
new_df = new_df.astype({'Kathmandu': float})
new_df
```

```python
new_df = new_df.interpolate()
new_df
```
