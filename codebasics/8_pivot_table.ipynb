```python
import pandas as pd
df = pd.read_csv("city_temp.csv")
df
```

```python
custom_df = pd.DataFrame({
    "day": [1, 2, 3, 1, 2, 3, 1, 2, 3],
    "city": ["new york", "new york", "new york", "mumbai", "mumbai", "mumbai", "beijing", "beijing", "beijing"],
    "temperature": [65, 66, 68, 75, 78, 82, 80, 77, 79],
    "humidity": [56, 58, 60, 80, 83, 85, 26, 30, 35],
})
```

```python
custom_df
```

```python
custom_df.pivot(index='day', columns='city')
```

```python
custom_df.pivot(index='day', columns='city', values="temperature")
```

# Pivot table

```python
custom_df.pivot_table(index='city', columns='day')
```

```python
custom_df.pivot_table(index='city', columns='day', aggfunc='sum')
```

```python
custom_df.pivot_table(index='city', columns='day', aggfunc='count')
```

```python
custom_df.pivot_table(index='city', columns='day') # Mean
```

```python
custom_df.pivot_table(index=pd.Grouper(freq='D', key='date'), columns='city')
```
